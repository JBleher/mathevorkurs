# Functions

## Examples for mathematical functions in R:

The function $f(x) = \frac{1}{x}$ can be defined in R with the function

```r
f <- function(x){
    return (1/x)
}
```

We can get the function values by handing a value $x$ to the function. Since a function is nothing else than a relation, for which for each $x$ value a $y$ value is defined -- we  können wir einfach eine Menge an $x$ Werten generieren, für die wir die Funktion auswerten wollen.

```r
x = c(2,3,1,0.001)
y = f(x)
print(y)
```

Since the function $f(x)$ is defined at each point, except for $x=0$, we give the functions all values $x\in\mathbb{R}\setminus 0$ as an input argument. If we plug $x_0 = 0$ into the function and evaluate $f$ at $x_0$ in R we get no error or warning, but the result is set to $\texttt{Inf}$ -- the value <a href="http://stat.ethz.ch/R-manual/R-devel/library/base/html/is.finite.html">R assigns to infinite numbers</a>.

```r
x0 = 0
f(x0)
```

In order to graph the function, we can simulate a quasi-continuous function by evaluating the function at points that are located very close to each other. With the function $\texttt{seq(start,stop,step)}$ we can easily generate a sequence of equidistant points.

```r
x = seq(-10,10,0.1)
y = f(x)
```

To make things more interesting, we can also look look at the function $g(x) = \sqrt{x}$.

```r
g <- function(x){
    return(sqrt(x))
}

z <- g(x)
```

Both functions $f$ and $g$ are now plotted into one graph.

```r
plot(x,y,type="l",col="black",xlab="x",ylab="y")
lines(x,z,col="blue")
legend("topleft", legend = c(expression(paste("f(x)=",frac(1,x))), expression(paste("g(x)=",sqrt(x)))),
               lty = c(1,1), col=c("black","blue"))
```

## Task 1

Plot the function $f(x) = x^3 - 3x^2+3x$.

```r
f <- function (x){
    return(x^3-3*x^2+3*x)
}

y <-f(x)
plot(x,y,type="l",col="black",xlab="x",ylab="y")
```

### Task 2:
Plot the function

\begin{equation*}
f(x)=
\begin{cases}
    5x - 2 & \text{for } x \lt 1 \\
    x^{2} + 4 & \text{for } x \geq 1
\end{cases}
\end{equation*}


Is the function continuous? If not, make it clear with a filled and an empty dot where the function is defined and where not. Moreover, make sure that the parts of the function which are not connected are displayed accordingly.

```r
f <- function (x){
    return(ifelse(x<1,5*x-2,x^2+4))
}
x <- seq(-1,2,3/1000)
y <- f(x)
plot(x[x<1],y[x<1],type="l",col="black",xlab="x",ylab="y",xlim=c(-1,2),ylim=c(-8,8))
lines(x[x>=1],y[x>=1],type="l",col="black")
points(1,3)
points(1,5,bg=c("black"),pch=21)
```

## Task 3
Plot a rational function with $n<m$. Also draw the asymptote.

```r
f <- function(x){
    return ((2*x - 1) / (x**2 + 1))
}
x <- seq(-20,20,40/1000)
y <- f(x)

plot(x,y,type="l",col="black",xlab="x",ylab="y")
abline(h=0,col="red")
```

## Task 4
Plot a rational function with $n>m$. Also draw the asymptote. Also make sure that poles are plotted correctly.

```r
f <- function(x){
   return((x**3 + 1) / (2*x - 1))
}
x <- seq(-20,20,40/10000)
y <- f(x)
plot(x,y,type="l",col="black",xlab="x",ylab="y",xlim=c(-20,20))
lines(x,0.5*x^2,col="red")
abline(v=0.5,col="blue")
```

## Task 5

Write a function $\text{comp}()$, which takes two functions as input and forms the composite.

```r
f <- function(x){
   return((x**3 + 1) / (2*x - 1))
}

g <- function(x){
    return ((2*x - 1) / (x**2 + 1))
}


comp <- function(f,g){
    function(...) f(g(...))
}

fg <- comp(f,g)
y <- fg(x)
plot(x,y,type="l",col="black",xlab="x",ylab="y",xlim=c(-20,20))
```

### Task 6
Plot both functions $f(x) = x^{2} - 1$  and  $g(x) = 2x - 1$ as well as the composite of both $f \circ g (x) = f(g(x))$.

```r
f <- function(x) x^2-1
g <- function(x) 2*x-1

fg <- comp(f,g)

y1 <- f(x)
y2 <- g(x)
y3 <- fg(x)

plot(x,y1,type="l",col="black",xlab="x",ylab="y",xlim=c(-10,10),ylim=c(-10,10))
lines(x,y2,col="red")
lines(x,y3,col="blue")
legend("bottomright", 
       legend = c(
           expression(paste("f(x)= ",x^2-1)), 
           expression(paste("g(x)= ",2*x -1)),
           expression(paste("f(g(x))= ",(2*x-1)^2-1))
       ),
               lty = c(1,1,1), col=c("black","blue","red"))
```
